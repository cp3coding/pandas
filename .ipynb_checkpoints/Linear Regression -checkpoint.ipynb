```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv', index = False)
df.head(2)
```

```python
df.rename(columns={'per capita income (US$)':'income'}, inplace = True)
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df.income,color='red',marker='+')
```

```python
new_df = df.drop('income',axis='columns')
new_df
```

```python
income = df.income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
new_df = pd.read_excel('year.xlsx', index_col= False)
```

```python
new_df
```

```python
p = reg.predict(new_df)
```

```python
p
```

```python
new_df['income']=p
```

```python
new_df
```

```python
new_df.to_excel("year.xlsx")
```
